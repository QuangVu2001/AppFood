```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 10):
        plt.plot(x, np.sin(x + i * .5) * (10 - i) * flip)
```

```python
graph = figure(title = 'Bokeh Chart')
sectors = ['Nam','Nu','LGBT']
percentage = percent_GT
radians = [math.radians((percent_GT/100)*360) for percent_GT in percentage]
```

```python
start_angle = [math.radians(0)]
prev = start_angle[0]
for i in radians[:-1]:
  start_angle.append(i + prev)
  prev = i + prev
end_angle = start_angle[1:] + [math.radians(0)]
```

```python
x=0
y=0
radius = 1
color = ['gold','silver','orange']
for i in range(len(sectors)):
    p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
        tools="hover", tooltips="@country: @value")

    p.wedge(x, y, radius,
                start_angle = start_angle[i],
                end_angle = end_angle[i],
                color = color[i],
                legend_label = sectors[i])
  

show(p)
```
